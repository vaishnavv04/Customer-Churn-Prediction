```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , StandardScaler
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data
```

```python
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
data
```

```python
data['Geography'].value_counts()
```

```python
label_encoder_gender = LabelEncoder()
data['Gender']= label_encoder_gender.fit_transform(data['Gender'])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder()
geo = onehot.fit_transform(data[['Geography']]).toarray()
```

```python
labels = onehot.get_feature_names_out(['Geography'])
```

```python
geo_df = pd.DataFrame(geo.toarray(),columns=labels)
```

```python
data.drop(['Geography'],axis=1,inplace=True)
```

```python
data = pd.concat([data,geo_df],axis=1)
```

```python
data
```

```python
# save the encoders
with open('label_encoder_gender.pkl','wb') as file :
    pickle.dump(label_encoder_gender,file)
with open('onehot_geo.pkl','wb') as file :
    pickle.dump(onehot,file)
```

```python
X = data.drop(['Exited'],axis=1)
y = data['Exited']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
with open('scaler.pkl','wb') as file :
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping , TensorBoard
import datetime
```

```python
model = Sequential([
    Dense(128,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(64,activation='relu'),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
log_dir = "logs/fit"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tf_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## set up earlystopping 
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,callbacks=[tf_callback,early_stopping_callback])
```

```python
model.save('model.h5')
```
